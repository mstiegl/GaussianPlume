# Gaussian Plume Modelling

## Point Source

```python
import numpy as np
import matplotlib.pyplot as plt
import gaussianPlume

%matplotlib inline
```

```python
rate = 1. # g/s/m2
H = 1. # m
U = 5. # m/s
xGrid = np.linspace(0.1 , 1000, 100) # m
yGrid = np.linspace(-500, 500, 100) # m
zGrid = 10. #m

pointSource = gaussianPlume.pointSource(0,0,0,rate,H)
grid = gaussianPlume.receptorGrid(xGrid,yGrid,zGrid)
stability = gaussianPlume.stabilityClass('C')

a = gaussianPlume.gaussianPlume(pointSource,grid,stability,U)

concField = a.calculateConcentration()*1e6

fig, ax = plt.subplots()
c = ax.contourf(grid.xMesh[0],grid.yMesh[0],concField[0], 1000)
cb = fig.colorbar(c)
cb.set_label('Concentration r(\$mu$ g/m$^3$)')
ax.set_xlabel('X [m]')
ax.set_ylabel('Y [m]')
```
